```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create some sample data
np.random.seed(10)
states = ['AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA']
dates = pd.date_range(start='2020-03-21', periods=10)
cases = np.random.randint(10, 1000, size=(10, 10))

# Create a DataFrame
data = pd.DataFrame(cases, columns=states, index=dates)

# Melt the DataFrame to long format
data_melted = data.melt(var_name='State', value_name='Cases')

# Bar plot
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x='State', y='Cases', data=data_melted, ax=ax)

# Customize the plot
ax.set_title('Bar Plot of COVID-19 Cases by State (Sample Data)')
ax.set_xlabel('State')
ax.set_ylabel('Cases')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create some sample data
np.random.seed(10)
states = ['AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA']
cases = np.random.randint(100, 1000, size=10)

# Create a DataFrame
data = pd.DataFrame({'State': states, 'Cases': cases})

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(data['Cases'], labels=data['State'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of COVID-19 Cases by State (Sample Data)')

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create some sample data
np.random.seed(10)
states = ['AK', 'AL', 'AR', 'AZ', 'CA']  # Select a few states for this example
dates = pd.date_range(start='2020-03-21', periods=10)
cases = np.random.randint(10, 1000, size=(10, 5))  # Adjust the number of states

# Create a DataFrame
data = pd.DataFrame(cases, columns=states, index=dates)

# Line chart
fig, ax = plt.subplots(figsize=(10, 6))
for state in states:
    plt.plot(data.index, data[state], label=state)

# Customize the plot
plt.title('Line Chart of COVID-19 Cases by State (Sample Data)')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.legend()  # Add legend to identify each state line
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```
